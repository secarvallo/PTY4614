## 2. Evaluación de la Situación

### Dataset Principal

- **Archivo:** `lung_cancer_data.csv`
- **Tamaño:** 300,000 pacientes
- **Features:** 30 variables (demográficas, clínicas, ambientales, outcomes)
- **Período:** Diagnósticos 2000-2024

**Categorías de Variables:**
- Demográficas: Age, Gender, Residential_Area
- Tabaquismo: Smoking_History, Pack_Years, Years_Smoked
- Ambientales: Air_Quality_Index, Exposure_to_Toxins
- Salud Personal: BMI, Physical_Activity_Level, Dietary_Habits
- Clínicas: Lung_Function_Test_Result, Chronic_Cough, Shortness_of_Breath
- Outcomes: Stage_of_Cancer, Survival_Years, Treatment_Type


## 3. Objetivos del Data Mining

### Tipo de Problema

**Clasificación Binaria Probabilística**
- Target: Estadio del cáncer (Early [I-II] vs Advanced [III-IV])
- Salida: Probabilidad de riesgo (0-100%) + nivel de riesgo

### Métricas de Rendimiento

| Métrica | Objetivo Mínimo | Objetivo Ideal | Justificación |
|---------|-----------------|----------------|---------------|
| **AUC-ROC** | ≥ 0.75 | ≥ 0.85 | Capacidad discriminativa buena a excelente |
| **Recall** | ≥ 0.70 | ≥ 0.80 | Minimizar falsos negativos (crítico en salud) |
| **Precision** | ≥ 0.65 | ≥ 0.75 | Reducir falsos positivos |
| **F1-Score** | ≥ 0.67 | ≥ 0.77 | Balance Precision-Recall |

> Objetivos mínimos son realistas para MVP académico. Ideales representan excelencia clínica.

### Algoritmos Candidatos

| Algoritmo | Rol | Justificación |
|-----------|-----|---------------|
| XGBoost | Modelo Principal | Máxima precisión, manejo de desbalanceo |
| Logistic Regression | Baseline | Interpretabilidad directa |
| Random Forest | Comparación | Robusto, feature importance |

## 4. Preguntas de Negocio (7 Total)

### TOP 3 PRIORITARIAS - SPRINT 1

#### Pregunta #1: Pack-Years x Edad

**¿A partir de cuántos "pack-years" de exposición al tabaco se incrementa significativamente el riesgo, y cómo varía según edad?**

- Variables: Pack_Years, Age, Smoking_History, Stage_of_Cancer
- Análisis: Curvas ROC por grupo de edad (18-40, 41-60, 61+)
- Resultado: Tabla de umbrales críticos por edad
- Tiempo: 2-3 días

#### Pregunta #2: Función Pulmonar + Síntomas

**¿Existe correlación entre función pulmonar y síntomas clínicos en la detección temprana?**

- Variables: Lung_Function_Test_Result, Chronic_Cough, Shortness_of_Breath, Chest_Pain_Symptoms
- Análisis: Matriz de correlación + Chi-square test
- Resultado: Sistema de alerta clínica (HIGH/MEDIUM/ROUTINE)
- Tiempo: 2-3 días

#### Pregunta #3: Urbano vs Rural

**¿Existe diferencia significativa en riesgo entre zonas urbanas vs rurales?**

- Variables: Residential_Area, Air_Quality_Index, Stage_of_Cancer
- Análisis: T-test/ANOVA + análisis descriptivo
- Resultado: Comparativa de riesgo + recomendaciones por zona
- Tiempo: 1-2 días

### Preguntas Complementarias 

4. **Probabilidad a 5 años:** Modelado de supervivencia o proxy temporal
5. **BMI y Actividad Física:** Feature importance en estadios tempranos
6. **Años de Cesación:** Análisis de ex-fumadores vs nunca fumadores
7. **Hábitos Protectores:** SHAP values para factores protectores

```python
# Configuración de entorno
import warnings
warnings.filterwarnings('ignore')

# Librerías base
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Configuración de visualización
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
pd.set_option('display.max_columns', None)

print("Librerías cargadas exitosamente")
print(f"Pandas version: {pd.__version__}")
print(f"NumPy version: {np.__version__}")
```

```python
# Cargar dataset
data_path = Path('../data/raw/lung-cancer-structured-clinical-dataset/lung_cancer_data.csv')

if data_path.exists():
    df = pd.read_csv(data_path)
    print(f"Dataset cargado exitosamente")
    print(f"Dimensiones: {df.shape[0]:,} filas x {df.shape[1]} columnas")
    print(f"Tamaño en memoria: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
else:
    print(f"ERROR: No se encontró el archivo en {data_path}")
```

```python
# Primeras filas del dataset
if 'df' in locals():
    display(df.head(10))
```

```python
# Información del dataset
if 'df' in locals():
    print("\nINFORMACIÓN DEL DATASET")
    print("=" * 80)
    df.info()
```

```python
# Distribución de Stage_of_Cancer
if 'df' in locals() and 'Stage_of_Cancer' in df.columns:
    print("\nDISTRIBUCIÓN DE ESTADIOS")
    print("=" * 80)
    
    stage_counts = df['Stage_of_Cancer'].value_counts().sort_index()
    stage_pct = (stage_counts / len(df)) * 100
    
    display(pd.DataFrame({'Count': stage_counts, 'Percentage': stage_pct}))
    
    # Gráfico
    plt.figure(figsize=(10, 5))
    stage_counts.plot(kind='bar', color=['#2ecc71', '#f39c12', '#e74c3c', '#c0392b'])
    plt.title('Distribución de Estadios de Cáncer', fontsize=14)
    plt.xlabel('Estadio')
    plt.ylabel('Número de Pacientes')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
    
    # Balance
    early = stage_counts.get('I', 0) + stage_counts.get('II', 0)
    advanced = stage_counts.get('III', 0) + stage_counts.get('IV', 0)
    print(f"\nEstadios Tempranos (I-II): {early:,} ({early/len(df)*100:.1f}%)")
    print(f"Estadios Avanzados (III-IV): {advanced:,} ({advanced/len(df)*100:.1f}%)")
    print(f"\nRatio Advanced/Early: {advanced/early:.2f}")
```
