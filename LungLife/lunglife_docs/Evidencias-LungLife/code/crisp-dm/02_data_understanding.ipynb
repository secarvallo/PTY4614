## 01.- Dataset: Lung Cancer Structured Clinical Dataset

### Descripcion del Dataset

Este dataset clinico estructurado contiene informacion comprehensiva sobre pacientes con cancer de pulmon. El cancer pulmonar es una enfermedad que se origina en los pulmones, frecuentemente asociada al tabaquismo, pero tambien vinculada a factores como la contaminacion del aire, predisposicion genetica y exposicion a toxinas como asbesto y radon.

### Contexto Clinico

| Aspecto | Descripcion |
|---------|-------------|
| **Tipos principales** | Cancer de Pulmon de Celulas No Pequenas (NSCLC) - mas comun; Cancer de Celulas Pequenas (SCLC) - mas agresivo |
| **Sintomas tipicos** | Tos persistente, dificultad respiratoria, dolor toracico, perdida de peso inexplicable |
| **Deteccion temprana** | Tomografias de baja dosis (LDCT) mejoran significativamente los resultados |
| **Tratamientos** | Cirugia, quimioterapia, radioterapia, inmunoterapia y terapias dirigidas |

### Relevancia para LungLife

Este dataset es fundamental para el MVP de LungLife porque:
- Contiene **300,002 registros** con 30 variables clinicas y demograficas
- Incluye informacion de **regiones chilenas prioritarias**: Antofagasta, Magallanes y Metropolitana

- Permite clasificacion binaria: **Early (Estadios I-II)** vs **Advanced (Estadios III-IV)**

## 4. Reporte de Calidad de Datos

### 4.1 Protocolo de Detección de Problemas

Este protocolo identifica sistemáticamente:
1. **Valores Nulos/Faltantes:** Impacto en entrenamiento del modelo
2. **Registros Duplicados:** Pueden sesgar métricas de evaluación
3. **Valores Atípicos (Outliers):** Afectan sensibilidad del modelo clínico
4. **Inconsistencias Lógicas:** Ej. Years_Smoked > Age

| Problema | Método de Detección | Umbral Crítico |
|----------|---------------------|----------------|
| Valores Nulos | `isnull().sum()` | > 5% por columna |
| Duplicados | `duplicated()` | > 1% del dataset |
| Outliers Numéricos | IQR × 1.5 / Z-score > 3 | Variable según contexto clínico |
| Inconsistencias | Reglas de negocio | Cualquier violación |

```python
# =============================================================================
# 4.2 ANÁLISIS DE VALORES NULOS Y FALTANTES
# =============================================================================

def analyze_missing_values(dataframe):
    """Genera un reporte detallado de valores faltantes."""
    missing_count = dataframe.isnull().sum()
    missing_pct = (missing_count / len(dataframe)) * 100
    
    missing_df = pd.DataFrame({
        'Variable': missing_count.index,
        'Valores_Faltantes': missing_count.values,
        'Porcentaje': missing_pct.values,
        'Tipo_Dato': [str(dataframe[col].dtype) for col in missing_count.index]
    })
    
    missing_df = missing_df.sort_values('Porcentaje', ascending=False)
    missing_df = missing_df[missing_df['Valores_Faltantes'] > 0]
    
    return missing_df

print("=" * 80)
print("ANÁLISIS DE VALORES FALTANTES")
print("=" * 80)

missing_report = analyze_missing_values(df)

if len(missing_report) == 0:
    print("\n[OK] EXCELENTE: No se detectaron valores faltantes en el dataset")
    print(f"   Total de registros verificados: {len(df):,}")
    print(f"   Total de variables verificadas: {len(df.columns)}")
else:
    print("\n[!] Se detectaron valores faltantes:")
    display(missing_report)
    
    # Visualización de valores faltantes
    if len(missing_report) > 0:
        fig, ax = plt.subplots(figsize=(10, 6))
        colors = ['#e74c3c' if x > 5 else '#f39c12' if x > 1 else '#2ecc71' 
                  for x in missing_report['Porcentaje']]
        bars = ax.barh(missing_report['Variable'], missing_report['Porcentaje'], color=colors)
        ax.set_xlabel('Porcentaje de Valores Faltantes (%)')
        ax.set_title('Distribución de Valores Faltantes por Variable')
        ax.axvline(x=5, color='red', linestyle='--', label='Umbral Crítico (5%)')
        plt.legend()
        plt.tight_layout()
        plt.show()

# Resumen ejecutivo
print("\n" + "-" * 40)
print("RESUMEN DE COMPLETITUD:")
print("-" * 40)
completeness = (1 - df.isnull().sum().sum() / (len(df) * len(df.columns))) * 100
print(f"Completitud del Dataset: {completeness:.2f}%")
print(f"Total de celdas: {len(df) * len(df.columns):,}")
print(f"Celdas con datos: {len(df) * len(df.columns) - df.isnull().sum().sum():,}")
```

```python
# =============================================================================
# 4.4 DETECCIÓN DE VALORES ATÍPICOS (OUTLIERS)
# =============================================================================

def detect_outliers_iqr(dataframe, columns):
    """Detecta outliers usando el método IQR (Interquartile Range)."""
    outlier_report = []
    
    for col in columns:
        if col in dataframe.columns and dataframe[col].dtype in ['int64', 'float64']:
            Q1 = dataframe[col].quantile(0.25)
            Q3 = dataframe[col].quantile(0.75)
            IQR = Q3 - Q1
            
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            
            outliers = dataframe[(dataframe[col] < lower_bound) | (dataframe[col] > upper_bound)]
            n_outliers = len(outliers)
            pct_outliers = (n_outliers / len(dataframe)) * 100
            
            outlier_report.append({
                'Variable': col,
                'Q1': Q1,
                'Q3': Q3,
                'IQR': IQR,
                'Límite_Inferior': lower_bound,
                'Límite_Superior': upper_bound,
                'N_Outliers': n_outliers,
                'Pct_Outliers': pct_outliers,
                'Min_Real': dataframe[col].min(),
                'Max_Real': dataframe[col].max()
            })
    
    return pd.DataFrame(outlier_report)

# Variables numéricas relevantes para el análisis clínico
clinical_numeric_vars = ['Age', 'Years_Smoked', 'Pack_Years', 'BMI', 
                         'Lung_Function_Test_Result', 'Air_Quality_Index',
                         'Tumor_Size_cm', 'Survival_Years', 'Follow_Up_Visits']

print("=" * 80)
print("DETECCIÓN DE VALORES ATÍPICOS (OUTLIERS)")
print("=" * 80)
print("Método: IQR (Interquartile Range) × 1.5")

outlier_df = detect_outliers_iqr(df, clinical_numeric_vars)
outlier_df = outlier_df.sort_values('Pct_Outliers', ascending=False)

print("\n[STATS] Resumen de Outliers por Variable:\n")
display(outlier_df[['Variable', 'Límite_Inferior', 'Límite_Superior', 
                    'Min_Real', 'Max_Real', 'N_Outliers', 'Pct_Outliers']].round(2))
```

## 5. Análisis de Desbalanceo de Clases

### 5.1 Importancia del Balance en Clasificación Clínica

En el contexto de detección temprana de cáncer pulmonar, el desbalanceo de clases tiene implicaciones críticas:

| Escenario | Consecuencia Clínica |
|-----------|---------------------|
| Sesgo hacia clase mayoritaria | Modelo no detecta casos Early → Pérdida de oportunidad diagnóstica |
| Sobre-representación de Advanced | Falsos negativos en estadios tempranos |
| Modelo optimizado solo para Accuracy | Ignora la clase minoritaria crítica |

**Objetivo:** Garantizar que el modelo tenga sensibilidad adecuada para detectar estadios tempranos (I-II).

```python
# =============================================================================
# 5.2 ANÁLISIS DETALLADO DEL DESBALANCEO
# =============================================================================

print("=" * 80)
print("ANÁLISIS DE DESBALANCEO DE CLASES")
print("=" * 80)

# Distribución por estadio original
stage_dist = df['Stage_of_Cancer'].value_counts().sort_index()
stage_pct = (stage_dist / len(df)) * 100

print("\n[STATS] DISTRIBUCION POR ESTADIO ORIGINAL:")
print("-" * 50)
for stage in ['I', 'II', 'III', 'IV']:
    if stage in stage_dist.index:
        count = stage_dist[stage]
        pct = stage_pct[stage]
        bar = "#" * int(pct // 2)
        print(f"  Estadio {stage}: {count:>8,} ({pct:5.1f}%) {bar}")

# Distribucion binaria (Early vs Advanced)
binary_dist = df['Cancer_Stage_Binary'].value_counts()
binary_pct = (binary_dist / len(df)) * 100

print("\n[STATS] DISTRIBUCION BINARIA (TARGET):")
print("-" * 50)
for stage in ['Early', 'Advanced']:
    count = binary_dist[stage]
    pct = binary_pct[stage]
    bar = "#" * int(pct // 2)
    color_indicator = "[E]" if stage == 'Early' else "[A]"
    print(f"  {color_indicator} {stage:10s}: {count:>8,} ({pct:5.1f}%) {bar}")

# Metricas de desbalanceo
n_early = binary_dist.get('Early', 0)
n_advanced = binary_dist.get('Advanced', 0)
majority = max(n_early, n_advanced)
minority = min(n_early, n_advanced)
imbalance_ratio = majority / minority if minority > 0 else float('inf')

print("\n[METRICS] METRICAS DE DESBALANCEO:")
print("-" * 50)
print(f"  - Clase Mayoritaria: {'Early' if n_early > n_advanced else 'Advanced'} ({majority:,})")
print(f"  - Clase Minoritaria: {'Advanced' if n_early > n_advanced else 'Early'} ({minority:,})")
print(f"  - Ratio de Desbalanceo: {imbalance_ratio:.2f}:1")
print(f"  - Proporcion Minority/Majority: {(minority/majority)*100:.1f}%")
```

```python
# =============================================================================
# 7.3 ANÁLISIS DE CORRELACIÓN CON LA VARIABLE OBJETIVO
# =============================================================================

print("=" * 80)
print("ANÁLISIS DE CORRELACIÓN CON LA VARIABLE OBJETIVO")
print("=" * 80)

# Preparar datos para correlación
# Codificar variables categóricas para análisis
df_encoded = df.copy()

# Mapeos para variables categóricas ordinales
ordinal_mappings = {
    'Physical_Activity_Level': {'Low': 0, 'Moderate': 1, 'High': 2},
    'Dietary_Habits': {'Poor': 0, 'Average': 1, 'Good': 2},
    'Medication_Response': {'Poor': 0, 'Moderate': 1, 'Good': 2},
    'Symptom_Progression': {'Worsening': 0, 'Stable': 1, 'Improving': 2}
}

for col, mapping in ordinal_mappings.items():
    if col in df_encoded.columns:
        df_encoded[f'{col}_encoded'] = df_encoded[col].map(mapping)

# Variables numéricas para correlación
numeric_for_corr = ['Age', 'Years_Smoked', 'Pack_Years', 'BMI', 
                    'Lung_Function_Test_Result', 'Air_Quality_Index',
                    'Tumor_Size_cm', 'Survival_Years', 'Follow_Up_Visits',
                    'Physical_Activity_Level_encoded', 'Dietary_Habits_encoded',
                    'Cancer_Stage_Numeric']

# Calcular correlaciones con el target
correlations_with_target = {}
for col in numeric_for_corr:
    if col in df_encoded.columns and col != 'Cancer_Stage_Numeric':
        valid_data = df_encoded[[col, 'Cancer_Stage_Numeric']].dropna()
        if len(valid_data) > 0:
            corr, p_value = spearmanr(valid_data[col], valid_data['Cancer_Stage_Numeric'])
            correlations_with_target[col] = {'correlation': corr, 'p_value': p_value}

# Crear DataFrame de correlaciones
corr_df = pd.DataFrame(correlations_with_target).T
corr_df = corr_df.sort_values('correlation', key=abs, ascending=False)
corr_df['abs_correlation'] = corr_df['correlation'].abs()
corr_df['significant'] = corr_df['p_value'] < 0.05

print("\n[STATS] Correlacion de Spearman con Cancer_Stage_Numeric:")
print("-" * 70)
display(corr_df.round(4))
```

```python
# =============================================================================
# 7.5 ANÁLISIS DE ASOCIACIÓN PARA VARIABLES CATEGÓRICAS
# =============================================================================

print("=" * 80)
print("ANÁLISIS DE ASOCIACIÓN: VARIABLES CATEGÓRICAS vs ESTADIO")
print("=" * 80)

categorical_features_to_test = [
    'Gender', 'Smoking_History', 'Residential_Area', 'Occupation',
    'Physical_Activity_Level', 'Dietary_Habits', 'Comorbidities',
    'Family_History_Cancer', 'Exposure_to_Toxins', 'Previous_Cancer_Diagnosis',
    'Chest_Pain_Symptoms', 'Shortness_of_Breath', 'Chronic_Cough', 'Weight_Loss'
]

chi2_results = []

for feature in categorical_features_to_test:
    if feature in df.columns:
        contingency = pd.crosstab(df[feature], df['Cancer_Stage_Binary'])
        chi2, p_value, dof, expected = chi2_contingency(contingency)
        
        # Calcular Cramér's V (medida de asociación)
        n = contingency.sum().sum()
        min_dim = min(contingency.shape[0] - 1, contingency.shape[1] - 1)
        cramers_v = np.sqrt(chi2 / (n * min_dim)) if min_dim > 0 else 0
        
        chi2_results.append({
            'Variable': feature,
            'Chi2': chi2,
            'p_value': p_value,
            'Cramers_V': cramers_v,
            'Significativo': 'Sí' if p_value < 0.05 else 'No'
        })

chi2_df = pd.DataFrame(chi2_results).sort_values('Cramers_V', ascending=False)

print("\n[STATS] Test Chi-cuadrado y Cramer's V:")
print("-" * 80)
display(chi2_df.round(4))

print("\n[INFO] INTERPRETACION DE CRAMER'S V:")
print("   - V < 0.1: Asociacion debil")
print("   - 0.1 <= V < 0.3: Asociacion moderada")
print("   - V >= 0.3: Asociacion fuerte")
```
